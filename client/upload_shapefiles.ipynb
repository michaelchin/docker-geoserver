```python
# zip the shapefiles. 
# The "-j" option is very important. 
# It will not create folders inside the zip file. It is important for geoserver.

import os, time
from pathlib import Path

Path('zipped-shapefiles').mkdir(parents=True, exist_ok=True)

data_dir='./Global_EarthByte_GPlates_PresentDay_ContinentalPolygons'

for path in Path(data_dir).rglob('*.shp'):
    print(path.stem)
    zip_fn = f'./zipped-shapefiles/{path.stem}.zip'
    os.system(f'zip -j {zip_fn} {data_dir}/{path.stem}.*')
    
    
```

```python
# upload shapefile and publish the layer

import os, time
from pathlib import Path

for path in Path('./zipped-shapefiles/').rglob('*.zip'):
    print(path.stem)
    print(r'{}'.format(path))
    
    ret=geo.create_shp_datastore(path=r'{}'.format(path), 
                         store_name=str(path.stem), workspace='Lithodat')
 
    print(ret)
    time.sleep(5)
```
